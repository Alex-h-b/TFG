### Goles


```python
import pandas as pd

# Unir datasets
full_data1 = pd.concat([df1[0], df2[0], df3[0], df4[0], df5[0]], ignore_index=True)
full_data1.sort_values(['Player', 'Season'], inplace=True)
# Eliminar columnas duplicadas manteniendo la primera aparición
full_data1 = full_data1.loc[:, ~full_data1.columns.duplicated()]

# Filtrar solo las columnas necesarias
features_goles1 = [
    # Volumen y edad
    "stats_Min", "stats_90s", "stats_Age",

    # Remate
    "shooting_Sh", "shooting_SoT", "shooting_SoT%", "shooting_G/Sh", "shooting_G/SoT",
    "shooting_Dist", "shooting_xG", "shooting_npxG", "shooting_npxG/Sh",
    "shooting_G-xG", "shooting_np:G-xG",

    # Creación
    "passing_xA", "passing_KP", "passing_PPA", "passing_CrsPA",
    "gca_SCA", "gca_GCA",

    # Participación ofensiva
    "possession_PrgC", "possession_Carries", "possession_Touches",
    "possession_Att Pen", "possession_Rec", "possession_PrgR",

    # Precisión pase
    "passing_Cmp%", "passing_Att", "passing_TotDist", "passing_PrgDist",

    # Categóricas
    "stats_Pos", "stats_Squad", "stats_Comp"
]

target1 = 'stats_Gls'

cols_keep1 = ['Player', 'Season'] + features_goles1 + [target1]
full_data1 = full_data1[cols_keep1]

# Crear features de la temporada anterior
for col in features_goles1:
    full_data1[f"{col}_prev"] = full_data1.groupby("Player")[col].shift(1)

# Variable objetivo actual
full_data1['target_Gls'] = full_data1[target1]

# Eliminar filas sin datos de temporada previa
full_data1.dropna(subset=[f"{col}_prev" for col in features_goles1], inplace=True)


```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# --- Definir variables base ---
features_goles1 = [
    'stats_Min', 'stats_90s', 'stats_Age', 'shooting_Sh', 'shooting_SoT', 'shooting_SoT%',
    'shooting_G/Sh', 'shooting_G/SoT', 'shooting_Dist', 'shooting_xG', 'shooting_npxG',
    'shooting_npxG/Sh', 'shooting_G-xG', 'shooting_np:G-xG', 'passing_xA', 'passing_KP',
    'passing_PPA', 'passing_CrsPA', 'gca_SCA', 'gca_GCA', 'possession_PrgC', 'possession_Carries',
    'possession_Touches', 'possession_Att Pen', 'possession_Rec', 'possession_PrgR',
    'passing_Cmp%', 'passing_Att', 'passing_TotDist', 'passing_PrgDist'
]
categorical_base1 = ['stats_Pos', 'stats_Squad', 'stats_Comp', "Player"]
target1 = 'stats_Gls'

# --- Filtrar y preparar datos ---
cols_keep1 = ['Season'] + features_goles1 + categorical_base1 + [target1]
full_data1 = full_data1[cols_keep1].copy()

# Convertir target a numérico (evita el error TypeError)
full_data1[target1] = pd.to_numeric(full_data1[target1], errors='coerce').fillna(0)

# Ordenar para hacer shift (temporada previa)
full_data1.sort_values(['Player', 'Season'], inplace=True)

# Crear variables previas (shift 1 temporada por jugador)
for col in features_goles1 + categorical_base1:
    full_data1[f"{col}_prev"] = full_data1.groupby('Player')[col].shift(1)

# Crear target
full_data1['target_Gls'] = full_data1[target1]

# Quitar filas sin datos previos
full_data1.dropna(subset=[f"{features_goles1[0]}_prev"], inplace=True)

# --- Identificar columnas ---
feature_cols_prev1 = [f"{col}_prev" for col in features_goles1 + categorical_base1]
categorical_cols1 = [col for col in feature_cols_prev1 if any(x in col for x in categorical_base1)]
numerical_cols1 = [col for col in feature_cols_prev1 if col not in categorical_cols1]

# --- Convertir datos ---
for col in numerical_cols1:
    full_data1[col] = pd.to_numeric(full_data1[col], errors='coerce').fillna(0)
for col in categorical_cols1:
    full_data1[col] = full_data1[col].astype(str).fillna('missing')

# --- Preparar X e y ---
X1 = full_data1[feature_cols_prev1]
y1 = full_data1['target_Gls']

# --- División train-test ---
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

# --- Pipeline con XGBoost ---
preprocessor_xgb = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols1),
        ('num', StandardScaler(), numerical_cols1)  # Escalar numéricas para XGBoost
    ],
    remainder='passthrough'
)

pipeline_xgb = Pipeline([
    ('preprocessor', preprocessor_xgb),
    ('model', XGBRegressor(objective='reg:squarederror', random_state=42))
])

# --- Parámetros para Grid Search ---
param_grid_xgb = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [3, 6, 9],
    'model__learning_rate': [0.01, 0.1],
    'model__gamma': [0, 0.1],
    'model__subsample': [0.8, 1.0]
}

# --- Grid Search ---
grid_search_xgb = GridSearchCV(
    estimator=pipeline_xgb,
    param_grid=param_grid_xgb,
    cv=5,
    scoring='neg_mean_absolute_error',
    n_jobs=-1,
    verbose=2
)

grid_search_xgb.fit(X_train1, y_train1)

# --- Resultados ---
print("\nMejores parámetros XGBoost:")
print(grid_search_xgb.best_params_)

best_model_xgb = grid_search_xgb.best_estimator_
y_pred_xgb = best_model_xgb.predict(X_test1)

print(f"\nMAE (XGBoost): {mean_absolute_error(y_test1, y_pred_xgb):.2f}")
print(f"R2 (XGBoost): {r2_score(y_test1, y_pred_xgb):.2f}")

# --- Importancia de características ---
importances_xgb = best_model_xgb.named_steps['model'].feature_importances_
feature_names_xgb = (
    best_model_xgb.named_steps['preprocessor']
    .named_transformers_['cat']
    .get_feature_names_out(categorical_cols1).tolist() +
    numerical_cols1
)

feat_imp_xgb = pd.DataFrame({'Características': feature_names_xgb, 'Importancia': importances_xgb})
feat_imp_xgb = feat_imp_xgb.sort_values('Importancia', ascending=False).head(15)

# --- Gráfico de importancia ---
plt.figure(figsize=(12, 8))
sns.barplot(x='Importancia', y='Características', data=feat_imp_xgb, palette='viridis')
plt.title('Top 15 carecterísticas más importantes')
plt.tight_layout()
plt.show()

# --- Predicción para nueva temporada ---
data_24_25 = full_data1[full_data1['Season'] == '2024-25'].copy()
X_pred = data_24_25[feature_cols_prev1]
data_24_25['goles_pred_xgb'] = best_model_xgb.predict(X_pred)

top_10_xgb = data_24_25.sort_values('goles_pred_xgb', ascending=False).head(10)[
    ['Player', 'stats_Squad', 'target_Gls', 'goles_pred_xgb']
]

print("\nTop 10 Predicciones 2025-26 (XGBoost):")
print(top_10_xgb)
top_10_xgb.to_csv('top_goles_predichos_xgboost.csv', index=False, float_format='%.1f')

```

### Asistencias

```python
import pandas as pd

# Unir datasets
full_data2 = pd.concat([df1[0], df2[0], df3[0], df4[0], df5[0]], ignore_index=True)
full_data2.sort_values(['Player', 'Season'], inplace=True)
# Eliminar columnas duplicadas manteniendo la primera aparición
full_data2 = full_data2.loc[:, ~full_data2.columns.duplicated()]

# Filtrar solo las columnas necesarias
features_asistencias2 = [
    # Volumen y edad
    "stats_Min", "stats_90s", "stats_Age",

    # Creación
    "passing_xA", "passing_KP", "passing_PPA", "passing_CrsPA", "passing_1/3",
    "passing_Ast",

    # Participación ofensiva
    "gca_SCA", "gca_GCA", "possession_Rec", "possession_Att Pen",
    "possession_PrgC", "possession_Carries", "possession_Touches",
    "possession_PrgR",

    # Precisión de pase
    "passing_Cmp%", "passing_Att", "passing_TotDist", "passing_PrgDist",

    # Categóricas
    "stats_Pos", "stats_Squad", "stats_Comp"
]

target2 = 'stats_Ast'

cols_keep2 = ['Player', 'Season'] + features_asistencias2 + [target2]
full_data2 = full_data2[cols_keep2]

# Crear features de la temporada anterior
for col in features_asistencias2:
    full_data2[f"{col}_prev"] = full_data2.groupby("Player")[col].shift(1)

# Variable objetivo actual
full_data2['target_Ast'] = full_data2[target2]

# Eliminar filas sin datos de temporada previa
full_data2.dropna(subset=[f"{col}_prev" for col in features_asistencias2], inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# --- Definir variables base ---
features_asistencias2 = [
    # Volumen y edad
    "stats_Min", "stats_90s", "stats_Age",

    # Creación
    "passing_xA", "passing_KP", "passing_PPA", "passing_CrsPA", "passing_1/3",
    "passing_Ast",

    # Participación ofensiva
    "gca_SCA", "gca_GCA", "possession_Rec", "possession_Att Pen",
    "possession_PrgC", "possession_Carries", "possession_Touches",
    "possession_PrgR",

    # Precisión de pase
    "passing_Cmp%", "passing_Att", "passing_TotDist", "passing_PrgDist",

    # Categóricas
    "stats_Pos", "stats_Squad", "stats_Comp", "Player"
]
categorical_base2 = ['stats_Pos', 'stats_Squad', 'stats_Comp', "Player"]
target2 = 'stats_Ast'

# --- Filtrar y preparar datos ---
cols_keep2 = ['Season'] + features_asistencias2 + [target2]
full_data2 = full_data2[cols_keep2].copy()

# Ordenar para hacer shift (temporada previa)
full_data2.sort_values(['Player', 'Season'], inplace=True)

# Asegurar que las columnas categóricas son strings y sin NaN
for col in categorical_base2:
    full_data2[col] = full_data2[col].astype(str).fillna('missing')

# Crear variables previas (shift 1 temporada por jugador)
# SOLO para las columnas originales, no las categóricas ya procesadas
for col in features_asistencias2:
    full_data2[f"{col}_prev"] = full_data2.groupby('Player')[col].shift(1)

# Eliminar filas con NaN en variables previas
full_data2.dropna(subset=[f"{col}_prev" for col in features_asistencias2], inplace=True)

# Crear target
full_data2['target_Ast'] = full_data2[target2]

# Identificar columnas categóricas y numéricas previas
# Usar las categóricas originales (no las _prev) para el encoding
feature_cols_prev2 = [f"{col}_prev" for col in features_asistencias2]
categorical_cols2 = categorical_base2  # Usar las categóricas originales
numerical_cols2 = [col for col in feature_cols_prev2 if col not in [f"{cat}_prev" for cat in categorical_base2]]

# Convertir numéricas (previas) a float
for col in numerical_cols2:
    full_data2[col] = pd.to_numeric(full_data2[col], errors='coerce').fillna(0)

# Preparar X e y - incluir categóricas originales y numéricas previas
X2 = full_data2[categorical_cols2 + numerical_cols2]
y2 = full_data2['target_Ast']

# División train-test
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Pipeline con preprocesamiento
preprocessor2 = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols2)
    ],
    remainder='passthrough'
)

pipeline2 = Pipeline([
    ('preprocessor', preprocessor2),
    ('model', RandomForestRegressor(random_state=42))
])

# Entrenar modelo
pipeline2.fit(X_train2, y_train2)

# Predecir y evaluar
y_pred2 = pipeline2.predict(X_test2)
mae2 = mean_absolute_error(y_test2, y_pred2)
r2_2 = r2_score(y_test2, y_pred2)

print(f"MAE: {mae2:.2f}")
print(f"R2: {r2_2:.2f}")
# --- Obtener nombres reales de las features tras la transformación ---
cat_feature_names = pipeline2.named_steps["preprocessor"] \
                             .named_transformers_["cat"] \
                             .get_feature_names_out(categorical_cols2)

# Combinar con las numéricas (que pasaron sin transformación)
all_feature_names = list(cat_feature_names) + numerical_cols2

# --- Emparejar nombres con importancias ---
importances = pipeline2.named_steps["model"].feature_importances_
feat_imp = list(zip(all_feature_names, importances))

# --- Ordenar por importancia descendente ---
feat_imp.sort(key=lambda x: x[1], reverse=True)

# --- Seleccionar top 15 ---
top_n = 15
top_features = feat_imp[:top_n]
names, imps = zip(*top_features)

# --- Graficar ---
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.barplot(x=imps, y=names, palette=sns.color_palette("viridis", len(imps)))
plt.title("Top 15 características más importantes")
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.tight_layout()
```
